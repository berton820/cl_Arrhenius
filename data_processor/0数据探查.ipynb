```python
# X = [5.968524065	,	5.829893169		,5.703601125	,	5.061009093		,4.373659225]
X = [390.92826	,340.32232	,299.9456	,157.74962	,79.3334 ]
Y = [2.302585093	,	0	,	-2.302585093	,	-4.605170186	,	-6.907755279]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 将X、Y转为numpy数组
X_np = np.array(X)
Y_np = np.array(Y)

# 最小二乘法拟合 y = kx + b
k, b = np.polyfit(X_np, Y_np, 1)

# 计算拟合直线的Y值
Y_fit = k * X_np + b

# 画出数据点和拟合直线
plt.scatter(X_np, Y_np, label='Data Points')
plt.plot(X_np, Y_fit, color='red', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('最小二乘法拟合')
plt.legend()
plt.show()

# 打印斜率和方程
print(f"斜率 k = {k}")
print(f"直线方程: y = {k:.4f}x + {b:.4f}")
```
