```python
import pandas as pd

file_path = '/Users/bertonyang/project/chenglu/data/TC4_0219.xlsx'

# 读取所有sheet
all_data = []
xlsx = pd.ExcelFile(file_path)
for sheet_name in xlsx.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
    temp = sheet_name  # 温度
    rates = df.iloc[0, 1:].reset_index(drop=True)  # 获取应变速率并重置索引，防止KeyError
    num_pairs = (df.shape[1] - 1) // 2  # 每两个列为一组，后面的是pair
    for i in range(num_pairs):
        strain_rate = rates.iloc[i * 2]  # 用iloc按位置索引，避免KeyError
        col_strain = 1 + i * 2
        col_stress = 2 + i * 2
        # 跳过完全是nan的列
        strain = df.iloc[1:, col_strain]
        stress = df.iloc[1:, col_stress]
        for s, r in zip(strain, stress):
            if pd.notna(s) and pd.notna(r):
                all_data.append({
                    '温度': temp,
                    '应变速率': strain_rate,
                    '真应变': s,
                    '流变应力': r
                })

# 转成DataFrame
processed_df = pd.DataFrame(all_data)
```
